```
dataframe = pandas.read_csv("all_buses_bf_9_99_999_only_fault_signales_training_testing.csv")

#convert the data into an array 
dataset = dataframe.values
feat=dataframe.keys()
feature_labels = feat.get_value
X = dataset[2:,1:23].astype(float)
Y = dataset[2:,22].astype(int)

#Split the data into trining and testing parts
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from skmultiflow.trees import HoeffdingTree
from sklearn.neural_network import MLPRegressor

import matplotlib.pyplot as plt
res = []

# Define a tree for fitting the complete dataset and one for streaming.
nn = MLPRegressor(max_iter=1000, activation='relu')
ht_partial = HoeffdingTree()
# Fit the complete dataset.
nn.fit(X_train, y_train)
nn_prediction = nn.predict(X_test)
mse=mean_squared_error(y_test,nn_prediction)
ht_complete_score = mse
print(f'Score when fitting at once: {ht_complete_score}')
# Streaming samples one after another.
timer = False
j = 0
for i in range(len(X_train)):
    ht_partial.partial_fit(X_train[i].reshape(1, -1), np.array([y_train[i]]))
    ht_prediction = ht_partial.predict(X_test)
    mse=mean_squared_error(y_test,ht_prediction)
    res.append(mse)
    print(f'Score when streaming after {i} samples: {res[-1]}')
    if res[-1] >= ht_complete_score - 0.01:
        print(f'(Almost) full score reached! Continue for another {50 - j} samples.')
        timer = True
    if timer:
        j += 1
        if j == 100:
           break

## NN
# Streaming samples one after another.
res_nn = []
nn2 = MLPRegressor(max_iter=1000, activation='relu', hidden_layer_sizes=(50,50,50,50,50))

timer = False
j = 0
for i in range(len(X_train)):
    nn2.partial_fit(X_train[i].reshape(1, -1), np.array([y_train[i]]))
    nn_prediction2 = nn2.predict(X_test)
    mse_nn=mean_squared_error(y_test,nn_prediction2)
    res_nn.append(mse_nn)
    print(f'Score when streaming after {i} samples: {res[-1]}')
    if res_nn[-1] >= ht_complete_score - 0.01:
        print(f'(Almost) full score reached! Continue for another {50 - j} samples.')
        timer = True
    if timer:
        j += 1
        if j == 100:
            break

#RF
res_rf = []
rf = RandomForestRegressor( criterion='mse', warm_start=True )

timer = False
j = 0
for i in range(len(X_train)):
    rf.fit(X_train[i].reshape(1, -1), np.array([y_train[i]]))
    rf_prediction = rf.predict(X_test)
    mse_rf=mean_squared_error(y_test,rf_prediction)
    res_rf.append(mse_nn)
    print(f'Score when streaming after {i} samples: {res[-1]}')
    if res_rf[-1] >= ht_complete_score - 0.01:
        print(f'(Almost) full score reached! Continue for another {50 - j} samples.')
        timer = True
    if timer:
        j += 1
        if j == 100:
            break


```

```

# Plot the scores after each sample.
plt.figure(figsize=(8, 6))
#plt.plot([0, i], [ht_complete_score, ht_complete_score], '--', label='Neural Network')
plt.plot(res_rf, label='Incrementally built RF')
plt.plot(res, label='Incrementally built Hoeffding Tree')
plt.plot(res_nn, label='Incrementally built Deep Neural Network')

plt.xlabel('Number of Samples', fontsize=15)
#plt.xticks(np.arange(len(X_train)))
plt.ylabel('MSE', fontsize=15)
plt.title('DNN vs HT vs RF ', fontsize=20)
plt.legend()
```
